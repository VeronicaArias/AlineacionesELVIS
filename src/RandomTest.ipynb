```python
n_points = 10000
phi = random.random(n_points)*2.*pi
theta = arccos(2.0*random.random(n_points)-1.0)
```

```python
x=cos(phi)*sin(theta)
y=sin(phi)*sin(theta)
z=cos(theta)
```

```python
scatter(x,y,s=0.1)
```

```python
n_axis = 42
axis_x = x[n_axis]
axis_y = y[n_axis]
axis_z = z[n_axis]
```

```python
cos_mu = x*axis_x + y*axis_y + z*axis_z
mu = arccos(cos_mu)
```

```python
a = hist(cos_mu)#ditribucion del coseno de los angulos
```

```python
a = hist(mu*180/pi)#distribucion de angulos
```
